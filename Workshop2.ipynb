Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python

def unique_list(lst):
    unique_lst = []
    for item in lst:
        if item not in unique_lst:
            unique_lst.append(item)
    return unique_lst

sample_list = [1, 2, 3, 3, 3,3, 4, 5]

unique_list_result = unique_list(sample_list)
print("Unique List :", unique_list_result)
```

Write a Python function that takes a number as a parameter and checks whether the number
is prime or not

```python

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

number = 17
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")

```

Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
def twin_primes(limit):
    twins = []
    for num in range(3, limit - 1, 2):
        if is_prime(num) and is_prime(num + 2):
            twins.append((num, num + 2))
    return twins
twin_prime_pairs = twin_primes(1000)
for pair in twin_prime_pairs:
    print(pair)
```

A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python

def sumOfDivisors(n):
    if n < 1:
        return 0

    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i

    return total
# Example usage:
n = 6
result = sumOfDivisors(n)
print(f"The sum of the divisors of {n} (excluding itself) is: {result}")
```

Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
Marks = [20, 30, 35, 55, 70, 80, 90]
# Function to calculate mean
def calculate_mean(marks):
    total = sum(marks)
    mean = total / len(marks)
    return mean
# Function to calculate median
def calculate_median(marks):


    sorted_marks = sorted(marks)
    n = len(sorted_marks)
    if n % 2 == 0:
        median = (sorted_marks[n // 2 - 1] + sorted_marks[n // 2]) / 2
    else:
        median = sorted_marks[n // 2]
    return median
mean = calculate_mean(Marks)
median = calculate_median(Marks)
print(f"Mean: {mean}")
print(f"Median: {median}")
```

Write a function that converts a decimal number to binary number

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n = n // 2
    return binary
# Example usage
decimal_number = 10
binary_number = decimal_to_binary(decimal_number)
print(f"The binary representation of {decimal_number} is {binary_number}.")

```

Write a function prodDigits() that inputs a number and returns the product of digits of that
number

```python
def prodDigits(n):
    product = 1

    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10

    return product
num = 1234
print("Product of digits:", prodDigits(num))
```
