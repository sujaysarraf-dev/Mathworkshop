Write a function squaredSum that returns the sum of the squares of an array/list of
integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python
to show the logic in your solution, do not use any library functions.
e. Write a function called sumDigits that is given an integer num and returns the sum of
the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if
there is one).
“The End”

```python
def myPow(x, n):
    result = 1
    count = 0
    while count < n:
       result = result * x
       count = count + 1
    return result
print(myPow(2, 3))
```

Write a function called sumDigits that is given an integer num and returns the sum of
the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if
there is one).
“The End”

```python
def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10
    return total
print(SumDigits(234))
```

```python
def squaredSum(numbers):
    total = 0
    for num in numbers:
        total = total + (num * num)
    return total
print(squaredSum([2, 3, 4]))
```

```python
def sumDigits(num):
    total = 0
    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10
    return total
print(sumDigits(123456))
```
